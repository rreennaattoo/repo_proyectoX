```python
import pandas as pd
import altair as alt

# Cargar datos, especificando el separador como punto y coma
df = pd.read_csv('/content/visualizacion base de datos blancura 2.csv', sep=';')

# Handle 'Calificación IMDB' column: replace comma with dot, convert to numeric, and drop NaNs
df['Calificación IMDB'] = df['Calificación IMDB'].str.replace(',', '.', regex=False)
df['Calificación IMDB'] = pd.to_numeric(df['Calificación IMDB'], errors='coerce')
df_cleaned = df.dropna(subset=['Calificación IMDB']).copy()

# 2️⃣ Dividir la columna de géneros en filas separadas
df_exploded = df_cleaned.assign(Género=df_cleaned['Género'].str.split(',')).explode('Género')
df_exploded['Género'] = df_exploded['Género'].str.strip()  # quitar espacios extra
df_exploded['Género'] = df_exploded['Género'].str.lower() # Convert to lowercase

# Handle 'comedia romántica' to be included in 'comedia' and 'romance'
def expand_genres(genres):
    expanded = []
    for genre in genres:
        expanded.append(genre)
        if 'comedia romántica' in genre:
            expanded.append('comedia')
            expanded.append('romance')
    return list(set(expanded)) # Use set to avoid duplicates and convert back to list

df_exploded['Género'] = df_exploded['Género'].apply(lambda x: expand_genres([x]))
df_exploded = df_exploded.explode('Género')
df_exploded = df_exploded[df_exploded['Género'] != 'comedia romántica'] # Remove the original 'comedia romántica' entry

# Map 'musical' to 'música'
df_exploded['Género'] = df_exploded['Género'].replace('musical', 'música')

# Map 'biografía' and 'cine histórico' to 'historia'
df_exploded['Género'] = df_exploded['Género'].replace(['biografía', 'cine histórico'], 'historia')

# Map 'thriller' to 'terror'
df_exploded['Género'] = df_exploded['Género'].replace('thriller', 'terror')

# Map 'suspenso' and 'misterio' to 'suspenso'
df_exploded['Género'] = df_exploded['Género'].replace(['suspenso', 'misterio'], 'suspenso')


# 3️⃣ Calcular el promedio de calificación por género
df_avg = df_exploded.groupby('Género', as_index=False)['Calificación IMDB'].mean()

# Capitalize the 'Género' for better visualization
df_avg['Género'] = df_avg['Género'].str.capitalize()


# 4️⃣ Crear el gráfico de barras con Altair
chart = alt.Chart(df_avg).mark_bar().encode(
    x=alt.X('Género:N', sort='-y', title='Género'),
    y=alt.Y('Calificación IMDB:Q', title='Promedio IMDb'),
    color=alt.Color('Género:N', scale=alt.Scale(scheme='category20')), # Use a different distinct color scheme
    tooltip=['Género', 'Calificación IMDB']
).properties(
    title='Promedio de calificación IMDb por género',
    width=600
)

chart
```
