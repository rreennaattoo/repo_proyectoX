```python
# 1. Importar librerías
import pandas as pd
import altair as alt

# 2. Cargar tu CSV (ajusta el nombre del archivo)
df = pd.read_csv('/content/base de datos visualizacion renato 2.2.2.csv', sep=';')

# 3. Reorganizar los datos en formato largo (long format)
# Inspect the columns to identify the correct names
print(df.columns)

# Assuming the correct columns are 'SEXO ACTOR PROTAGONISTA' and 'SEXO ACTOR SECUNDARIO '
df_long = pd.DataFrame({
    'rol': ['Principal'] * len(df) + ['Secundario'] * len(df),
    'genero': list(df['SEXO ACTOR PROTAGONISTA']) + list(df['SEXO ACTOR SECUNDARIO'])
})

# 4. Agrupar por género y rol
conteo = df_long.groupby(['rol', 'genero']).size().reset_index(name='cantidad')

# 5. Crear el gráfico
chart = alt.Chart(conteo).mark_bar().encode(
    x=alt.X('rol:N', title='Rol del Actor'),
    y=alt.Y('cantidad:Q', stack='normalize', title='Proporción de Género'),
    color=alt.Color('genero:N', title='Género del Actor', scale=alt.Scale(domain=['F', 'M'], range=['pink', 'blue'])), # Setting custom colors for F and M
    tooltip=['rol', 'genero', 'cantidad']
).properties(
    title='Distribución de género entre actores principales y secundarios',
    width=400,
    height=300
)

chart
```

```python
import altair as alt

# Assuming 'chart' is the name of your Altair chart object

# Guardar como JPG
chart.save('distribucion_genero.jpg', scale_factor=2)

# Guardar como HTML
chart.save('distribucion_genero.html')

print("El gráfico ha sido guardado como 'distribucion_genero.jpg' y 'distribucion_genero.html'")
```

```python
import altair_saver

# Exportar el gráfico a HTML (interactivo)
chart.save('Distribución de género entre actores principales y secundarios.html')

# Exportar el gráfico a PNG (imagen estática)
altair_saver.save(chart, 'Distribución de género entre actores principales y secundarioso.png', method='vl-convert')

print("✅ Visualización guardada como HTML y PNG")
```

```python
import json
import vl_convert as vlc

# Assuming 'chart' is your Altair chart object
chart_json = chart.to_json()

# Define output filename
output_filename_jpg = 'distribucion_genero.jpg'

# Convert the Vega-Lite JSON to JPG
with open('chart.vl.json', 'w') as f:
    f.write(chart_json)

# Use VlConverter class to convert
converter = vlc.VlConverter()
with open(output_filename_jpg, 'wb') as f:
    f.write(converter.vegalite_to_bytes(chart_json, format='jpg', scale=2))


print(f"El gráfico ha sido guardado como '{output_filename_jpg}'")
```
