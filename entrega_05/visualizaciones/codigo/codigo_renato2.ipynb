## Preparar Datos de Género de Directores

### Subtask:
Se analizará la columna 'director-01' para identificar el género (masculino/femenino) de cada director y se creará una nueva columna 'genero_director'. Luego, se agruparán los datos por año y género para contar el número de directores y directoras por cada año.


**Reasoning**:
The first step is to inspect the 'director-01' column to understand its content and format, which is crucial for inferring gender later. Displaying unique values will help identify common naming patterns and potential challenges.



```python
print(df['director-01'].value_counts())
```

**Reasoning**:
As per the subtask instructions, a new column named 'genero_director' needs to be created in the DataFrame, and it should be initialized with a default value. 'Desconocido' is a suitable default value as it indicates that the gender is yet to be determined.



```python
df['genero_director'] = 'Desconocido'
print("Columna 'genero_director' creada e inicializada.")
```

**Reasoning**:
To classify the gender of directors, I need to define lists of common male and female Spanish names. I will create a function that attempts to determine the gender based on the first name of each director. This function will then be applied to the 'director-01' column to populate the 'genero_director' column, marking names not found in either list as 'Desconocido'.



```python
nombres_masculinos = [
    'José', 'Juan', 'Luis', 'Carlos', 'Francisco', 'Manuel', 'Antonio', 'Miguel', 'Pedro', 'Pablo',
    'Fernando', 'Alejandro', 'Andrés', 'David', 'Jorge', 'Daniel', 'Javier', 'Sergio', 'Ricardo', 'Alberto',
    'Eduardo', 'Enrique', 'Rafael', 'Diego', 'Gonzalo', 'Héctor', 'Mario', 'Martín', 'Oscar', 'Ramón',
    'Roberto', 'Rubén', 'Santiago', 'Sebastián', 'Vicente', 'Esteban', 'Álvaro', 'Gabriel', 'Guillermo', 'Ignacio',
    'Joaquín', 'Leonardo', 'Marcos', 'Nicolás', 'Raúl', 'Rodrigo', 'Adrián', 'Agustín', 'Alfredo', 'Benjamín',
    'Cristián', 'Felipe', 'Germán', 'Gustavo', 'Patricio', 'Víctor'
]
nombres_femeninos = [
    'María', 'Ana', 'Laura', 'Sofía', 'Valentina', 'Camila', 'Isabel', 'Paula', 'Daniela', 'Carolina',
    'Andrea', 'Elena', 'Gabriela', 'Jimena', 'Lucía', 'Mariana', 'Natalia', 'Pilar', 'Sara', 'Silvia',
    'Verónica', 'Alejandra', 'Beatriz', 'Carmen', 'Clara', 'Diana', 'Dolores', 'Emilia', 'Eva', 'Florencia',
    'Gloria', 'Irene', 'Julia', 'Lorena', 'Magdalena', 'Marta', 'Mónica', 'Patricia', 'Raquel', 'Rocío',
    'Rosa', 'Sandra', 'Teresa', 'Victoria', 'Ximena', 'Cecilia', 'Consuelo', 'Francisca', 'Jessica', 'Katherine',
    'Claudia', 'Elizabeth', 'Macarena', 'Melissa', 'Romina', 'Soledad'
]

def inferir_genero(nombre_completo):
    if pd.isna(nombre_completo):
        return 'Desconocido'
    # Convertir a minúsculas para una comparación sin distinción de mayúsculas y minúsculas
    nombre_completo_lower = str(nombre_completo).lower()

    # Separar el nombre en palabras y tomar la primera como posible nombre propio
    primer_nombre = nombre_completo_lower.split(' ')[0]

    # Buscar nombres femeninos
    for nombre_fem in nombres_femeninos:
        if primer_nombre == nombre_fem.lower():
            return 'Femenino'

    # Buscar nombres masculinos
    for nombre_masc in nombres_masculinos:
        if primer_nombre == nombre_masc.lower():
            return 'Masculino'

    return 'Desconocido'

df['genero_director'] = df['director-01'].apply(inferir_genero)

print("La columna 'genero_director' ha sido actualizada con el género inferido.")
print(df['genero_director'].value_counts())
```

**Reasoning**:
Now that the 'genero_director' column has been populated, the next step is to group the DataFrame by 'año' and 'genero_director' and then count the number of unique directors for each group, as specified in the subtask. This will provide the annual count of male and female directors.



```python
directores_por_ano_genero = df.groupby(['año', 'genero_director'])['director-01'].nunique().reset_index()

print("Conteo de directores únicos por año y género:")
print(directores_por_ano_genero.head())
```

## Calcular la Tasa de Mujeres por Hombre

### Subtask:
Utilizando los conteos de directores y directoras por año, se calculará la tasa de mujeres por hombre (F/M) para cada año.


**Reasoning**:
To calculate the female-to-male ratio, I need to reshape the `directores_por_ano_genero` DataFrame so that each year has separate columns for 'Femenino', 'Masculino', and 'Desconocido' director counts. Pivoting will achieve this, and filling `NaN` values with 0 ensures that years without a particular gender count are handled correctly for subsequent calculations.



```python
if 'directores_por_ano_genero' in locals():
    directores_pivot = directores_por_ano_genero.pivot_table(
        index='año',
        columns='genero_director',
        values='director-01',
        fill_value=0
    ).reset_index()

    # Ensure all gender columns exist, filling with 0 if not present in the pivot
    # Now, we expect only 'Femenino' and 'Masculino' columns.
    for col in ['Femenino', 'Masculino']:
        if col not in directores_pivot.columns:
            directores_pivot[col] = 0

    print("DataFrame pivotado por año y género (solo Femenino/Masculino):")
    print(directores_pivot.head())
else:
    print("No se ha definido 'directores_por_ano_genero', no se puede pivotar el DataFrame.")
```

**Reasoning**:
Now that the data is pivoted, I need to calculate the female-to-male ratio. This requires dividing the 'Femenino' column by the 'Masculino' column, and it's crucial to handle cases where 'Masculino' is zero to avoid division errors and replace resulting infinite or NaN values with 0 to ensure the ratio is well-defined.



```python
if 'directores_pivot' in locals():
    directores_pivot['ratio_F_M'] = directores_pivot['Femenino'] / directores_pivot['Masculino']
    directores_pivot['ratio_F_M'] = directores_pivot['ratio_F_M'].replace([float('inf'), -float('inf')], 0).fillna(0)

    print("DataFrame con la tasa de mujeres por hombres (solo Femenino/Masculino):")
    print(directores_pivot.head())
else:
    print("No se ha definido 'directores_pivot', no se puede calcular la tasa F/M.")
```

## Crear Gráfico de Área Apilado y Línea

### Subtask:
Se generará un gráfico de Altair que combine un gráfico de área apilado para mostrar la cantidad de directores y directoras a lo largo de los años, superponiendo un gráfico de línea con un eje secundario para visualizar la tasa de mujeres por hombre.


**Reasoning**:
To prepare the data for the stacked area chart, I need to 'melt' the `directores_pivot` DataFrame. This will transform the gender columns ('Femenino', 'Masculino', 'Desconocido') into rows, making the data suitable for Altair's stacking capabilities.



```python
import altair as alt

df_melted = directores_pivot.melt(
    id_vars=['año'],
    value_vars=['Femenino', 'Masculino'], # Se excluye 'Desconocido' aquí
    var_name='genero_director',
    value_name='count'
)

print("DataFrame 'df_melted' creado y listo para la visualización (sin 'Desconocido'):")
print(df_melted.head())
```

**Reasoning**:
Now that the data is prepared in `df_melted` for the stacked area chart and `directores_pivot` for the ratio line chart, I will create both charts, combine them using `alt.layer`, assign appropriate axes, tooltips, and titles, and then display the final visualization as per the subtask instructions.



```python
base = alt.Chart(df_melted).encode(
    x=alt.X('año:O', axis=alt.Axis(title='Año')),
    tooltip=['año', 'genero_director', 'count']
)

# Gráfico de área apilado para el conteo de directores por género
area_chart = base.mark_area().encode(
    y=alt.Y('count:Q', axis=alt.Axis(title='Número de Directores')),
    color=alt.Color('genero_director:N', title='Género del Director',
                    scale=alt.Scale(domain=['Femenino', 'Masculino'], range=['#88937B', '#5A0F05'])) # Colores actualizados
)

# Gráfico de línea para la tasa F/M
line_chart = alt.Chart(directores_pivot).mark_line(color='red').encode(
    x=alt.X('año:O'),
    y=alt.Y('ratio_F_M:Q', axis=alt.Axis(title='Ratio Femenino/Masculino', titleColor='red')),
    tooltip=['año', alt.Tooltip('ratio_F_M', format='.2f', title='Ratio F/M')]
)

# Combinar los gráficos
combined_chart = alt.layer(area_chart, line_chart).resolve_scale(
    y='independent'  # Permite que ambos gráficos tengan sus propios ejes Y
).properties(
    title='Evolución de la Participación de Directores por Género y Ratio F/M en el Cine Chileno'
).interactive()

combined_chart
```

## Visualización de los 20 Directores con Más Películas

### Subtask:
Se identificará a los 20 directores con la mayor cantidad de películas y se creará un gráfico de barras horizontal que muestre este listado, utilizando una paleta de colores personalizada.

**Reasoning**:
To identify the top 20 directors, I need to count the number of unique films associated with each director from the `df` DataFrame. Grouping by 'director-01' and counting 'nombre-pelicula' will provide this, and then sorting and selecting the top 20 will give the required data for the visualization.

```python
directores_conteo = df.groupby('director-01')['nombre-pelicula'].nunique().reset_index()
directores_conteo = directores_conteo.rename(columns={'nombre-pelicula': 'cantidad_peliculas'})

top_20_directores = directores_conteo.sort_values(by='cantidad_peliculas', ascending=False).head(20)

print("Los 20 directores con la mayor cantidad de películas:")
print(top_20_directores)
```

**Reasoning**:
Now that I have the data for the top 20 directors, I will create an Altair horizontal bar chart. The chart will display the director names on the Y-axis and the count of films on the X-axis.
```python
import altair as alt

palette = ["#41402A", "#5A0F05", "#9B4337", "#88937B", "#CFC38D", "#778F3B", "#598299"]

chart_top_directors = alt.Chart(top_20_directores).mark_bar().encode(
    x=alt.X('cantidad_peliculas:Q', title='Cantidad de Películas'),
    y=alt.Y('director-01:N', sort='-x', title='Director'),
    color=alt.Color('director-01:N', scale=alt.Scale(range=palette), legend=None), # Color by director using the palette
    tooltip=['director-01', 'cantidad_peliculas']
).properties(
    title='Top 20 Directores con Mayor Cantidad de Películas'
).interactive()

chart_top_directors
```
