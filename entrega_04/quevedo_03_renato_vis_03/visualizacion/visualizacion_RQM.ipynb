```python
import pandas as pd

# Cargar datos
df = pd.read_csv("/content/Base de datos final .csv", sep=";")


# Echar un vistazo
df.head(5)
```

```python
df_seleccion = df[["nombre-pelicula", "director-01"]]
df_seleccion.head()
```

```python
import altair as alt

# Count the number of movies per director
director_counts = df_seleccion['director-01'].value_counts().reset_index()
director_counts.columns = ['director', 'numero_peliculas']

# Sort by number of movies in descending order and take the top N directors for better readability
top_directors = director_counts.sort_values('numero_peliculas', ascending=False).head(20)

# Define the custom color scheme
color_scheme = ['#5A0F05', '#241D14', '#41402A']

# Create the bar chart with custom colors
chart = alt.Chart(top_directors).mark_bar().encode(
    x=alt.X('numero_peliculas', title='Number of Movies'),
    y=alt.Y('director', sort='-x', title='Director'),
    # Use the custom color scheme, cycling through them for each director
    color=alt.Color('director', scale=alt.Scale(range=color_scheme), legend=None),
    tooltip=['director', 'numero_peliculas']
).properties(
    title='Top 20 Directors by Number of Movies'
)

chart
```

## Generate Pie Chart

### Subtask:
Create an Altair pie chart to visualize the proportion of Male, Female, and Unknown/Other directors. Apply the specified color palette ('#5A0F05', '#241D14', '#41402A') to these categories. The chart should include labels for each slice indicating the gender category and its percentage, and have a clear title.


**Reasoning**:
I need to convert the aggregated gender counts into a pandas DataFrame, then use Altair to create a pie chart as specified, including custom colors, labels, percentages, and a title.



```python
import altair as alt
import pandas as pd

# Convert the aggregated_genders dictionary to a DataFrame
gender_df = pd.DataFrame(aggregated_genders.items(), columns=['Gender', 'Count'])

# Define the custom color scheme
color_scheme = ['#5A0F05', '#241D14', '#41402A']

# Create the base chart
base = alt.Chart(gender_df).encode(
    theta=alt.Theta("Count", stack=True)
)

# Create the pie chart (arcs)
pie = base.mark_arc(outerRadius=120).encode(
    color=alt.Color("Gender", scale=alt.Scale(range=color_scheme)),
    order=alt.Order("Count", sort="descending"),
    tooltip=["Gender", alt.Tooltip("Count", format=",.0f"), alt.Tooltip("Count", format=".1%", title="Percentage")]
)

# Create the text labels for the percentages
text = base.mark_text(radius=140).encode(
    text=alt.Text("Count", format=".1%"),
    order=alt.Order("Count", sort="descending"),
    color=alt.value("black") # Set text color to black for better contrast
)

# Combine the pie chart and the text labels
chart = (pie + text).properties(
    title='Proportion of Directors by Inferred Gender'
)

chart
```
